```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Cannot open webcam")
    exit(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Finnished or Error")

        break

    cv2.imshow("Video Frame",frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)       # First curve (sine wave)
y2 = np.cos(x)       # Second curve (cosine wave)
y3 = np.sin(x) * 0.5 # Third curve (damped sine wave)

# Create figure
# plt.figure(figsize=(10, 6))

# Plot multiple curves on the same graph
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.plot(x, y3, label='0.5*sin(x)', color='green', linestyle=':', linewidth=2)

# Add titles and labels
# plt.title("Multiple Plots on Same Graph", fontsize=14, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)

# Add legend and grid
# plt.legend(fontsize=10)
# plt.grid(True, linestyle='--', alpha=0.5)

# Adjust layout and show plot
# plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 20)  # 20 points from 0 to 10
y = np.sin(x) + np.random.normal(0, 0.1, 20)  # Sin wave with some noise

# Create the customized plot
plt.plot(x, y, color='purple', linestyle='-.', marker='s')  
plt.title("Customized Line Plot")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.legend(["Trend"])
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title("Category-wise Values")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.savefig("plot.png")
plt.show()
```

```python
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])

# Add text at coordinates (x=1, y=10)
plt.text(1, 10, 'This is some text', 
         fontsize=12, 
         color='red',
         bbox=dict(facecolor='yellow', alpha=0.5))

# Add another text with rotation
plt.text(2, 5, 'Rotated text', 
         rotation=45,
         fontstyle='italic')

plt.title('Simple Text Example')
plt.savefig("plot.png")
plt.show()
```

#pie chart

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',       # Show percentages
    startangle=90,           # Start from top
    explode=[0, 0.1, 0, 0],  # Emphasize one slice
    shadow=True              # Fixed: moved this parameter inside parentheses
)
plt.title("Enhanced Pie Chart")
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Use ggplot style
plt.style.use('ggplot')

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot with ggplot style
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Wave (ggplot style)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.savefig("plot.png")
plt.show()
```
